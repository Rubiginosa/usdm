```python
from google.colab import files
uploaded = files.upload()
import torchaudio  # or librosa, etc.

audio_path = list(uploaded.keys())

```

```python
import whisper

model = whisper.load_model("turbo")
for audio_path in list(uploaded.keys()):
  result = model.transcribe(audio_path)
  print(result["text"])
  with open(f'{audio_path}.transcript', 'w') as f:
    f.write(result["text"])
```
