```python
ground_truth = "Please call Stella.  Ask her to bring these things with her from the store:  Six spoons of fresh snow peas, five thick slabs of blue cheese, and maybe a snack for her brother Bob."

def normalize(string):
  string = string.lower()
  string = ''.join(char for char in string if char.isalnum() or char.isspace())
  string = string.replace('ok', 'okay')
  return string
```

```python
normalize(ground_truth)
```

```python
scores = {
    'afrikaans': (0,0),
    'english': (0,0),
    'hindi': (0,0),
    'spanish': (0,0),
    'korean': (0,0),
}
for file in os.listdir('/content/drive/MyDrive/cs-436-final/speech archive'):
  if os.path.splitext(os.fsdecode(file))[1] =='.user_text':
    #print(os.fsdecode(file))
    with open(f'/content/drive/MyDrive/cs-436-final/speech archive/{file}', 'r') as f:
      content = f.read()
      score = jiwer.wer(ground_truth, normalize(content))
      print(file, score)



```
